```python
from tensorflow.keras.preprocessing import image
import numpy as np
```

```python
img = image.load_img('10002.jpg', target_size=(224, 224))
```

```python
img_array = image.img_to_array(img)
```

```python
print(img_array.shape)
```

```python
expanded_img_array = np.expand_dims(img_array, axis=0)
```

```python
print(expanded_img_array.shape)
```

```python
expanded_img_array
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
import tensorflow
from tensorflow.keras.preprocessing import image
from tensorflow.keras.layers import GlobalMaxPooling2D
import numpy as np
```

```python
model = ResNet50(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)

# Freeze model
model.trainable = False

# Add pooling layer
model = tensorflow.keras.Sequential([
    model,
    GlobalMaxPooling2D()
])
```

```python
preprocessed_img = preprocess_input(expanded_img_array)

print(preprocessed_img.shape)

```

```python
preprocessed_img
```

```python
model.predict(preprocessed_img).shape
```

```python
model.predict(preprocessed_img).flatten()
```

```python
model.predict(preprocessed_img).flatten() / np.linalg.norm(model.predict(preprocessed_img).flatten())
```

```python
np.sqrt(np.dot(model.predict(preprocessed_img).flatten(), model.predict(preprocessed_img).flatten()))
```
